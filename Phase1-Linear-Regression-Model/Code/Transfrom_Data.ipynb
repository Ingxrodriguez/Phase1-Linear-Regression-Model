# Fixing Categorical Variables 

Lest use dummies to convert the categorical variables into numerical variables. I  use the get_dummies() method to convert the categorical variables into numerical variables(one-hot encoding) is a way to convert categorical variables (labels or groups) into numeric columns 1 and 0 or Bolean values True and False. This is useful for machine learning algorithms that can only work with numerical data. The get_dummies() method takes the data frame and the columns that we want to convert into dummy variables. The drop_first=True parameter is used to drop the first column of the dummy variables to avoid multicollinearity.

```python
#A function that converts categorical values into new binary columns—one for each different category

Clean_df = pd.get_dummies(df, columns=['NdsCH', 'NdssC'], drop_first=True)

Clean_df.head()

   
```
