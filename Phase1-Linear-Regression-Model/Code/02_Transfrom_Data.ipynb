## Data transformation 

In the transformation stage, I  Started importing the neccesary  Libraries. Then I loaded my Data Set, I checked my raw data and the and I rename the columns to identfy the right features. So I started look for duplicates observations and missing values to ensure the quality of data. One duplicate row was identified and removed, while no missing values were found in the dataset. Additionally, the variables NdsCH and NdssC were identified as categorical variables because they contained a limited number of unique integer values. To make these variables suitable for linear regression, they were converted into binary indicator variables using one-hot encoding with pd.get_dummies(). Finally, the cleaned dataset was saved as a new CSV file so it can be used directly in the modeling stage without repeating the transformation steps.

# Fixing Categorical Variables 

Lest use dummies to convert the categorical variables into numerical variables. I  use the get_dummies() method to convert the categorical variables into numerical variables(one-hot encoding) is a way to convert categorical variables (labels or groups) into numeric columns 1 and 0 or Bolean values True and False. This is useful for machine learning algorithms that can only work with numerical data. The get_dummies() method takes the data frame and the columns that we want to convert into dummy variables. The drop_first=True parameter is used to drop the first column of the dummy variables to avoid multicollinearity.

```python
#A function that converts categorical values into new binary columns—one for each different category

Clean_df = pd.get_dummies(df, columns=['NdsCH', 'NdssC'], drop_first=True)

Clean_df.head()

   
```
